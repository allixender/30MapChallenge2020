```python
x = """Õpperada
Matkarada
Rattarada
Maastikusõidurada
Telkimisala
Lõkkekoht
Kattega lõkkekoht
Metsaonn
Metsamaja
Looduskeskus
Külastuskeskus
Teabepunkt
Vaatetorn
Puhkekoht
Matkavõimalus
Prügi sorteerimise võimalus
Telkimisvõimalus
Invavõimalus
Talvine kasutus
Muu"""


y = """Study trail
Hiking trail
Cycling trail
Off-road trail
Campsite
Campfire site
Outdoor fireplace
Forest hut
Rental hut
Nature House
Visitor Centre
Information point
Observation Tower
Rest Stop
Hiking options
Recycling point
Camping facilities
Wheelchair accessibility
Winter accessibility
Other"""
```

```python
x1 = x.split('\n')
x1
```

```python
y1 = y.split('\n')
```

```python
b = [(a[0],a[1]) for a in zip(x1,y1)]
```

```python
b
```

```python
def transla(ee):
    for tup in b:
        if tup[0].lower() == ee.lower():
            return tup[1].lower()
        
gdf['type_eng'] = gdf['type'].apply(transla)
```

```python
gdf
```

```python
gdf.loc[gdf['type_eng'].isna()]
```

```python
gdf['type'].unique()
```

```python
gdf['type_eng'].unique()
```

```python
poi3301 = gdf.to_crs(epsg=3301)
```

```python
fig, ax = plt.subplots(figsize=(20,15))

plt.title('Estonian State Forest Management Centre (RMK) public recreations sites',fontsize=20)
plt.rc('legend', fontsize=16)

est3301.plot(facecolor='black', edgecolor='black', lw=0.7, alpha=0.8, ax=ax)
poi3301.plot(column='type_eng', ax=ax, categorical=True, legend=True, markersize=11, cmap='tab20')

# Adjust legend location
leg = ax.get_legend()
leg.set_bbox_to_anchor((1.0,0.66))



ax.set_axis_off()

plt.tight_layout()
fig.savefig('../source/_static/day-01-points.png', bbox_inches='tight')
```
